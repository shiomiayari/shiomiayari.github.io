```python
!pip install pytesseract pillow

# ãƒ©ã‚¤ãƒ–ãƒ©ãƒªã‚¤ãƒ³ãƒãƒ¼ãƒˆ
import pytesseract
from PIL import Image
from google.colab import files
import os
import re
import cv2
import numpy as np

# ç”»åƒã‚’ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰
uploaded = files.upload()

# ã‚¢ãƒƒãƒ—ãƒ­ãƒ¼ãƒ‰ãƒ•ã‚¡ã‚¤ãƒ«ã®åå‰
filename = list(uploaded.keys())[0]

# ç”»åƒã‚’é–‹ã
img = Image.open(filename)


# TESSDATA_PREFIXç’°å¢ƒå¤‰æ•°ã‚’è¨­å®š
# Google Colabã§ã¯ã€tessdataãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒªã¯/usr/share/tesseract-ocr/4.00/tessdataã«ã‚ã‚Šã¾ã™ã€‚
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata'

# æ—¥æœ¬èªè¨€èªãƒ‡ãƒ¼ã‚¿ã‚’ã‚¤ãƒ³ã‚¹ãƒˆãƒ¼ãƒ«
!sudo apt install tesseract-ocr-jpn

# OCRã‚’å®Ÿè¡Œï¼ˆæ—¥æœ¬èªå¯¾å¿œï¼‰
text = pytesseract.image_to_string(img, lang="jpn")

# çµæœã‚’è¡¨ç¤º
print("æŠ½å‡ºã•ã‚ŒãŸãƒ†ã‚­ã‚¹ãƒˆ:\n", text)


#ã“ã“ã‹ã‚‰æ­£è¦è¡¨ç¾
import re
pattern = r"(\d{1,2}/\d{1,2})\s+(\d{1,2}:\d{2})?\s*(.+)"
matches = re.findall(pattern, text)

# æŠ½å‡ºçµæœã‚’è¡¨ç¤º
print("\nğŸ”¹ äºˆå®šä¸€è¦§:")
for match in matches:
    print(f"ğŸ“… æ—¥ä»˜: {match[0]}, ğŸ•’ æ™‚åˆ»: {match[1] if match[1] else 'çµ‚æ—¥'}, ğŸ“Œ äºˆå®š: {match[2]}")
```
