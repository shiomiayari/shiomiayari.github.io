```python
!pip install pytesseract pillow

# ライブラリインポート
import pytesseract
from PIL import Image
from google.colab import files
import os
import re
import cv2
import numpy as np

# 画像をアップロード
uploaded = files.upload()

# アップロードファイルの名前
filename = list(uploaded.keys())[0]

# 画像を開く
img = Image.open(filename)


# TESSDATA_PREFIX環境変数を設定
# tessdataディレクトリは/usr/share/tesseract-ocr/4.00/tessdata
os.environ['TESSDATA_PREFIX'] = '/usr/share/tesseract-ocr/4.00/tessdata'

# 日本語言語データ
!sudo apt install tesseract-ocr-jpn

# OCRを実行
text = pytesseract.image_to_string(img, lang="jpn")

# 結果表示
print("抽出されたテキスト:\n", text)


#正規表現
import re
pattern = r"(\d{1,2}/\d{1,2})\s+(\d{1,2}:\d{2})?\s*(.+)"
matches = re.findall(pattern, text)

# 抽出結果を表示
print("\n🔹 予定一覧:")
for match in matches:
    print(f"📅 日付: {match[0]}, 🕒 時刻: {match[1] if match[1] else '終日'}, 📌 予定: {match[2]}")

#正規表現
import re
pattern = r"(\d{1,2}月\d{1,2})日\s+(\d{1,2}:\d{2})?\s*(.+)"
matches = re.findall(pattern, text)

# 抽出結果を表示
print("\n🔹 予定一覧:")
for match in matches:
    print(f"📅 日付: {match[0]}, 🕒 時刻: {match[1] if match[1] else '終日'}, 📌 予定: {match[2]}")
```
